### भाग 6: Train-Validation Split (ट्रेन-व्हॅलिडेशन विभाजन)

```python
dataset_size = len(dataset)
train_size = int(0.8 * dataset_size)
train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)
```

**मराठी स्पष्टीकरण:**
- `dataset_size` - एकूण किती batches आहेत
- `train_size = 0.8 * dataset_size` - 80% batches ट्रेनिंगसाठी
- `dataset.take(train_size)` - पहिले 80% batches ट्रेनिंग सेट
- `dataset.skip(train_size)` - उरलेले 20% batches व्हॅलिडेशन सेट

### भाग 7: Performance Optimization (परफॉरमन्स ऑप्टिमायझेशन)

```python
AUTOTUNE = tf.data.AUTOTUNE
train_dataset = train_dataset.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_dataset = val_dataset.cache().prefetch(buffer_size=AUTOTUNE)
```

**मराठी स्पष्टीकरण:**
- `.cache()` - डेटा मेमरी मध्ये स्टोअर करतो (वेगाने लोड होते)
- `.shuffle(1000)` - 1000 इमेजेस रँडमली शफल करतो (ऑर्डर बदलतो)
- `.prefetch()` - पुढचा batch लोड करतो (मॉडेल प्रशिक्षण घेत असतानाच)
- `AUTOTUNE` - टेन्सरफ्लो ऑटोमॅटिकली बेस्ट सेटिंग्ज निवडतो

### भाग 9: Data Augmentation (डेटा ऑगमेंटेशन)

```python
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
])
```

**मराठी स्पष्टीकरण:**
- `RandomFlip` - इमेज उलटी-सुलटी करतो (आरशातील प्रतिमा)
- `RandomRotation(0.2)` - इमेज 20% पर्यंत फिरवतो
- `RandomZoom(0.2)` - इमेज 20% पर्यंत झूम करतो
- **उद्देश:** मर्यादित डेटा पासून नवीन वैरायटी तयार करणे (ओव्हरफिटिंग टाळते)

## 📗 दुसरा Notebook: Model Building (मॉडेल तयार करणे)

### भाग 1: CNN Model Architecture (सीएनएन मॉडेल आर्किटेक्चर)

```python
model = models.Sequential([
    data_augmentation,
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(2, activation='softmax')
])
```

**मराठी स्पष्टीकरण - प्रत्येक लेयर:**

**1. Data Augmentation Layer:**
- डेटा ऑगमेंटेशन लागू करतो

**2. Conv2D(32, (3,3), activation='relu'):**
- `32` - 32 फिल्टर्स (फीचर डिटेक्टर)
- `(3,3)` - फिल्टर साइज (3×3 pixels)
- `activation='relu'` - नॉन-लिनियरिटी जोडते (निगेटिव नंबर 0 करते)
- **काम:** इमेज मधून एज, कॉर्नर, टेक्सचर डिटेक्ट करते

**3. MaxPooling2D((2,2)):**
- 2×2 विंडो मधून मॅक्सिमम व्हॅल्यू निवडते
- **काम:** इमेजचा साइज अर्धा करते (128→64, 64→32, 32→16)
- **फायदा:** कॉम्प्युटेशन कमी होते, ओव्हरफिटिंग कमी होते

**4. Flatten():**
- 2D (16×16×128) ला 1D (32768) मध्ये बदलते
- **काम:** कन्व्होल्यूशन लेयर नंतर डेन्स लेयर ला कनेक्ट करते

**5. Dense(128, activation='relu'):**
- फुली कनेक्टेड लेयर (128 न्यूरॉन्स)
- **काम:** फीचर्स वरून क्लास डिसिजन घेते

**6. Dropout(0.5):**
- 50% न्यूरॉन्स रँडमली बंद करते
- **काम:** ओव्हरफिटिंग टाळते

**7. Dense(2, activation='softmax'):**
- आउटपुट लेयर (2 न्यूरॉन्स - दोन क्लासेस)
- `softmax` - प्रोबॅबिलिटी वितरीत करते (एकूण = 1)
- **उदा:** [0.9, 0.1] म्हणजे 90% क्लास 0, 10% क्लास 1

### भाग 2: Model Compilation (मॉडेल कंपाइल करणे)

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

**मराठी स्पष्टीकरण:**
- `optimizer='adam'` - वजन अपडेट करण्याची अल्गोरिदम (सर्वात लोकप्रिय)
  - काम: त्रुटी कमी करण्यासाठी weights बदलते
- `loss='sparse_categorical_crossentropy'` - एरर कॅल्क्युलेट करते
  - का? मल्टी-क्लास क्लासिफिकेशन साठी (2+ क्लासेस)
- `metrics=['accuracy']` - अक्युरसी मोजते

### भाग 3: Training (प्रशिक्षण)

```python
EPOCHS = 20
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=EPOCHS,
    callbacks=[early_stopping]
)
```

**मराठी स्पष्टीकरण:**
- `EPOCHS = 20` - संपूर्ण डेटासेट वर 20 वेळा प्रशिक्षण
- `fit()` - मॉडेल ट्रेन करते
- `train_dataset` - ट्रेनिंग डेटा
- `validation_data` - व्हॅलिडेशन डेटा (प्रत्येक epoch नंतर चेक करते)
- `callbacks=[early_stopping]` - जर सुधारणा थांबली तर लवकर थांबवते

**Training Process (समजून घ्या):**
```
Epoch 1: Model शिकते (अक्युरसी 40%)
Epoch 2: थोडे सुधारते (50%)
Epoch 3: अजून सुधारते (60%)
.
.
Epoch 15: अक्युरसी 95% पर्यंत पोहोचते
```

### भाग 4: Early Stopping (लवकर थांबवणे)

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

**मराठी स्पष्टीकरण:**
- `monitor='val_loss'` - व्हॅलिडेशन लॉस बघतो
- `patience=5` - जर 5 epochs मध्ये सुधारणा नाही झाली तर थांबवते
- `restore_best_weights=True` - जेव्हा लॉस सर्वात कमी होता तेव्हाचे weights परत आणते
- **उद्देश:** ओव्हरफिटिंग आणि वेळ वाया जाण्यापासून वाचवते

- `BATCH_SIZE = 16` - एकाच वेळी 16 इमेजेस प्रोसेस केल्या जातील (मेमरी कमी वापरते)

**BATCH_SIZE 16 का?** - 16 चा batch small आहे, त्यामुळे:
- कमी मेमरी लागते
- मॉडेल अधिक स्थिर (stable) प्रशिक्षण घेते
- प्रत्येक batch नंतर weights अपडेट होतात

## 📂 तिसरा भाग: Data Generators (डेटा जनरेटर)

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    horizontal_flip=True,
    zoom_range=0.2,
    fill_mode='nearest'
)
```

**मराठी अर्थ:**
- `rescale=1./255` - पिक्सल व्हॅल्यू 0-255 वरून 0-1 मध्ये बदलते (नॉर्मलायझेशन)
- `validation_split=0.2` - 20% डेटा व्हॅलिडेशनसाठी, 80% ट्रेनिंगसाठी
- `rotation_range=20` - इमेजेस 20 अंशापर्यंत फिरवतो (ऑगमेंटेशन)
- `horizontal_flip=True` - इमेजेस आडव्या दिशेने उलटतो (मिरर इमेज)
- `zoom_range=0.2` - इमेजेस 20% पर्यंत झूम करतो
- `fill_mode='nearest'` - रोटेशन/झूम नंतर रिकाम्या जागा nearest पिक्सेलने भरतो

### Training Generator:

```python
train_generator = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    subset='training',
    classes=['Pepper_bell_Bacterial_spot', 'Pepper_bell_healthy'],
    shuffle=True
)
```

**मराठी अर्थ:**
- `flow_from_directory` - फोल्डरमधून थेट इमेजेस लोड करते
- `target_size=(128,128)` - सर्व इमेजेस 128x128 मध्ये resize
- `class_mode='sparse'` - लेबल्स 0,1,2... अशा integer स्वरूपात
- `subset='training'` - हा ट्रेनिंग डेटा आहे (80% डेटा)
- `classes=[...]` - फक्त हे दोन फोल्डर वापरा
- `shuffle=True` - प्रत्येक epoch नंतर इमेजेस रँडमली मिसळा

## ⚖️ चौथा भाग: Class Weights (क्लास वेट्स)

## 🏗️ पाचवा भाग: CNN Model Architecture (सीएनएन मॉडेल आर्किटेक्चर)

### Block 1 (पहिला ब्लॉक):
```python
layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
layers.BatchNormalization(),
layers.MaxPooling2D((2,2)),
layers.Dropout(0.25),
```

**मराठी अर्थ:**
- `Conv2D(32, (3,3))` - 32 फिल्टर्स, प्रत्येक 3×3 साइजचा. **हे काय करते?** - इमेजमधून edges, corners, textures शोधते
- `activation='relu'` - नकारात्मक संख्या 0 करते. **का?** - non-linearity आणते (नेटवर्कला complex patterns शिकायला मदत)
- `BatchNormalization()` - प्रत्येक बॅच नंतर आउटपुट normalize करते. **फायदा:** त्वरीत training, स्थिरता
- `MaxPooling2D((2,2))` - 2×2 च्या विंडोमधून सर्वात मोठी व्हॅल्यू निवडते. **काय होते?** - इमेजचा साइज अर्धा होतो (128→64)
- `Dropout(0.25)` - 25% न्यूरॉन्स रँडमली बंद करते. **उद्देश:** ओव्हरफिटिंग टाळणे

### Block 2 (दुसरा ब्लॉक):
```python
layers.Conv2D(64, (3,3), activation='relu'),
layers.BatchNormalization(),
layers.MaxPooling2D((2,2)),
layers.Dropout(0.25),
```
- आता 64 फिल्टर्स (मागे 32 होते)
- इमेजचा साइज: 64×64 वरून 32×32

### Block 3 (तिसरा ब्लॉक):
```python
layers.Conv2D(128, (3,3), activation='relu'),
layers.BatchNormalization(),
layers.MaxPooling2D((2,2)),
layers.Dropout(0.25),
```
- 128 फिल्टर्स (जास्त फीचर्स शिकते)
- इमेजचा साइज: 32×32 वरून 16×16

### Global Pooling:
```python
layers.GlobalAveragePooling2D(),
```
- प्रत्येक फीचर मॅपची सरासरी काढते
- **फायदा:** ओव्हरफिटिंग कमी करते, पॅरामीटर्स कमी करते

### Dense Layers (फुली कनेक्टेड लेयर्स):
```python
layers.Dense(128, activation='relu'),
layers.Dropout(0.5),
layers.Dense(64, activation='relu'),
layers.Dropout(0.3),
layers.Dense(2, activation='softmax')
```
- `Dense(128)` - 128 न्यूरॉन्स, सर्व फीचर्स एकत्र करतात
- `Dropout(0.5)` - 50% न्यूरॉन्स बंद (मजबूत regularization)
- `Dense(64)` - 64 न्यूरॉन्स
- `Dense(2, activation='softmax')` - आउटपुट लेयर, 2 क्लासेस. `softmax` probability वितरीत करते (एकूण = 1)

**Model Parameters:**
```
Total params: 119,042 (465 KB)
Trainable params: 118,594 (463 KB)
```
- सुमारे 1.19 लाख पॅरामीटर्स
- हे मध्यम आकाराचे model आहे (खूप मोठे नाही, खूप लहान नाही)

## 🔧 सहावा भाग: Compile Model (मॉडेल कंपाइल करणे)

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

**मराठी अर्थ:**
- `optimizer='adam'` - weights update करण्याची अल्गोरिदम.
  - **Adam म्हणजे काय?** - Adaptive Moment Estimation. ही सर्वात लोकप्रिय optimizer आहे.
  - **कसे काम करते?** - प्रत्येक पॅरामीटरसाठी स्वतंत्र learning rate ठेवते
  
- `loss='sparse_categorical_crossentropy'` - एरर कॅल्क्युलेट करण्याची पद्धत.
  - **स्पार्स म्हणजे काय?** - लेबल्स integer स्वरूपात आहेत (0,1) one-hot नाहीत
  - **क्रॉसएन्ट्रोपी म्हणजे काय?** - predicted probability आणि actual label मधील फरक मोजते
  - उदा: actual=1, predicted=0.9 → loss कमी; actual=1, predicted=0.1 → loss जास्त

- `metrics=['accuracy']` - प्रशिक्षण दरम्यान अचूकता मोजत राहते

## 📞 सातवा भाग: Callbacks (कॉलबॅक्स)

```python
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3),
    tf.keras.callbacks.ModelCheckpoint('best_plant_disease_model.h5', monitor='val_accuracy', save_best_only=True)
]
```

**मराठी अर्थ:**

### 1. EarlyStopping (लवकर थांबवणे):
- `monitor='val_loss'` - व्हॅलिडेशन लॉस बघत राहा
- `patience=5` - जर 5 epochs मध्ये सुधारणा नाही झाली तर थांबा
- `restore_best_weights=True` - सर्वात चांगले weights परत आणा (जेव्हा loss सर्वात कमी होता)
- **उद्देश:** ओव्हरफिटिंग टाळणे आणि वेळ वाचवणे

### 2. ReduceLROnPlateau (लर्निंग रेट कमी करणे):
- `monitor='val_loss'` - व्हॅलिडेशन लॉस बघा
- `factor=0.5` - learning rate अर्धी करा (उदा. 0.001 → 0.0005)
- `patience=3` - जर 3 epochs मध्ये सुधारणा नाही झाली तर LR कमी करा
- **उद्देश:** minima मध्ये अडकून न राहण्यासाठी

### 3. ModelCheckpoint (मॉडेल सेव्ह करणे):
- `'best_plant_disease_model.h5'` - फाइल नाव
- `monitor='val_accuracy'` - व्हॅलिडेशन अक्युरसी बघा
- `save_best_only=True` - फक्त सर्वात चांगले model सेव्ह करा
- **उद्देश:** सर्वोत्तम model नंतर वापरण्यासाठी सेव्ह करणे
